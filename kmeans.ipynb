# Mall Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("Mall_Customers_Segmentation.csv")
df
```

```python
x=df.iloc[:,[3,4]].values
```

```python
np.random.seed(2)
from sklearn.cluster import KMeans
wcss=[]
for i in range (1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',n_init='auto')
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.xlabel("number of clusters")
plt.ylabel("wcss")
plt.show()
```

```python
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(x)
```

```python
np.random.seed(3)
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=5,init='k-means++',n_init='auto')
kmeans.fit(x)
r=kmeans.labels_
plt.scatter(x[:,0],x[:,1],s=100,c=r)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='red')
```

# Engg College Data

```python
df=pd.read_csv("Engg_Colleges_Data.csv")
```

## Convert string to categorical data

```python
for column,content in df.items():
    if(pd.api.types.is_string_dtype(content)):
        df[column]=content.astype("category").cat.as_ordered()
```

```python
for column,content in df.items():
    if not (pd.api.types.is_numeric_dtype(content)):
        df[column]=pd.Categorical(content).codes+1
```

```python
x=df.iloc[:,[1,4]].values
x
```

```python
np.random.seed(4)
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',n_init='auto')
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
```

```python
from sklearn.cluster import KMeans
k=KMeans(n_clusters=4,random_state=0,init='k-means++')
k.fit(x)
r=k.labels_
r
```

```python
plt.scatter(x[:,0],x[:,1],s=100,c=r)
```

# Kmedioids

```python
from sklearn_extra.cluster import KMedoids
k=KMedoids(n_clusters=4,random_state=0,init='heuristic')
k.fit(x)
r=k.labels_
r
```

```python
plt.scatter(x[:,0],x[:,1],s=100,c=r)
```
