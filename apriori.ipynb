```python
import pandas as pd
from apyori import apriori
```

```python
df=pd.read_csv("market.csv")
df
```

```python
oht = pd.get_dummies(df[['Bread', 'Honey', 'Bacon', 'Toothpaste', 'Banana', 'Apple', 'Hazelnut',
       'Cheese', 'Meat', 'Carrot', 'Cucumber', 'Onion', 'Milk', 'Butter',
       'ShavingFoam', 'Salt', 'Flour', 'HeavyCream', 'Egg', 'Olive', 'Shampoo',
       'Sugar']], columns=['Bread', 'Honey', 'Bacon', 'Toothpaste', 'Banana', 'Apple', 'Hazelnut',
       'Cheese', 'Meat', 'Carrot', 'Cucumber', 'Onion', 'Milk', 'Butter',
       'ShavingFoam', 'Salt', 'Flour', 'HeavyCream', 'Egg', 'Olive', 'Shampoo',
       'Sugar'])
oht
```

```python
from apyori import apriori
rules = apriori(transactions = oht, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2,use_colnames=True)
results=list(rules)
```

```python
def inspect(results):
    lhs=[tuple(result[2][0][0])[0] for result in results]
    rhs=[tuple(result[2][0][1])[0] for result in results]
    support=[result[1] for result in results]
    confidence=[result[2][0][2] for result in results]
    lift=[result[2][0][3] for result in results]
    return list(zip(lhs,rhs,support,confidence,lift))
result_new=pd.DataFrame(inspect(results),columns=['lhs','rhs','support','confidence','lift'])
result_new.sort_values(by="lift",ascending=False)[:10]
```
