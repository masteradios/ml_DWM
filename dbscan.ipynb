```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("species-Segmentation-dataset.csv")
df
```

```python
x=df.values[:,[2,3]]
x
```

```python
from sklearn.cluster import DBSCAN
dbscan=DBSCAN(eps=0.01,min_samples=3)
dbscan.fit(x)
r=dbscan.labels_
r
```

```python
plt.scatter(x[:,0],x[:,1],s=100,c=r)
plt.show()
```
